```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the Lena image
img = cv2.imread('Lena.png', cv2.IMREAD_GRAYSCALE)

# Resize the image to 4 different sizes
sizes = [128, 64, 32, 16]
resized_images = [cv2.resize(img, (size, size)) for size in sizes]

# Plot the original image and down sampled images
plt.subplot(2, 3, 1), plt.imshow(img, cmap='gray')
plt.title("Original Image")

for i, resized_img in enumerate(resized_images): #returns the index of each item in the list
    plt.subplot(2, 3, i + 2), plt.imshow(resized_img, cmap='gray')
    plt.title(f"Size: {sizes[i]}x{sizes[i]}") # sets the title of the plot to the size of the current resized image

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the Lena image
image = cv2.imread('Lena.png', cv2.IMREAD_GRAYSCALE)

# Create a list of intensity ranges
intensity_ranges = [255, 128, 64, 32, 16, 8, 4, 2]

# Create a list of down-sampled images
downsampled_images = []
for i in intensity_ranges:
    # Normalize the image to the given intensity range
    image_norm = cv2.normalize(image, None, 0, i-1, cv2.NORM_MINMAX)

    # Down-sample the image with cv2.resize()
    downsampled_image = cv2.resize(image_norm, (128, 128), interpolation=cv2.INTER_AREA)

    # Append the down-sampled image to the list
    downsampled_images.append(downsampled_image)

# Plot the original image and the down-sampled images in the same size display area
plt.figure(figsize=(12, 12))
plt.subplot(3, 3, 1)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Original Image')
for i in range(8):
    plt.subplot(3, 3, i + 2)
    plt.imshow(downsampled_images[i], cmap='gray')
    plt.axis('off')
    plt.title('Intensity range: 0 - ' + str(intensity_ranges[i]))
plt.show()
```
